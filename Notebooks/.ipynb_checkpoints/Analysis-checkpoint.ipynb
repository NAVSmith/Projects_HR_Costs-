```python
def anal_by_category(df):
    grouped_n_avg_df = group_n_add_cost_avg(df, ['Task_Category'])
    grouped_n_avg_df['Time_p%_4m_total'] = round(100 * grouped_n_avg_df.Time_Decimal / sum(grouped_n_avg_df.Time_Decimal),2)
    grouped_n_avg_df['Cost_p%_4m_total'] =  round(100 * grouped_n_avg_df.Task_Cost / sum(grouped_n_avg_df.Task_Cost),2)
    return grouped_n_avg_df
```

## deep diving into the devops component
let do an analysis by projects to check for outliers in the data.
 we can see that there is an outlier in the data for the DevOps the FinTech projects that it was 100% DevOps work
#compare the devlopment-- the data is to small for a deep dive but one can see the diffrance in range
See what we can get from that

```python
def anal_by_category_by_project(df):
    test=group_n_add_cost_avg(df, ['Project','Task_Category'])
    # gernerate a new array that is same shape an leangh of the orginal of 
    #and costistes of the total time spent on each project
    test1 =test.copy().groupby('Project').Time_Decimal.sum()[test.Project].reset_index().drop(columns=["Project"])
    # append this total arry as a new column to the df 
    test['Total_time_project'] =  test1
    # creates an new columns that is the % of the paricular task catergory time form the total time spent on the project
    test['Task_Time_p%_4m_total'] = round(100 *   test.Time_Decimal / test.Total_time_project,2)
    # repeats on for the Task costs
    test2 =test.copy().groupby('Project').Task_Cost.sum()[test.Project].reset_index().drop(columns=["Project"])
    test['Total_cost_project'] =  test2
    test['Task_cost_p%_4m_total'] = round(100 *   test.Task_Cost / test.Total_cost_project,2)
    # drops the midway columns
    return test.drop(columns=['Total_time_project', 'Total_cost_project'])
    # There is a more efficient way or shorter to do this use np.arrays but considering the small scope of the data it is ok.
```

```python
deep_dive_sma10 = anal_by_category_by_project(sma10_exp_pro)
deep_dive_sma10
```

```python
def add_seniority(df):
    df['seniority'] = np.where(df.Position.str.contains('Senior|Lead', regex=True), 'Senior', 'Junior')
    df['work'] = (np.where(df.Position.str.contains('Back', regex=True), 'Back End',  
             np.where(df.Position.str.contains('Front', regex=True), 'Front End',
             np.where(df.Position.str.contains('Full', regex=True), 'Full-Stack',
             np.where(df.Position.str.contains('Devops', regex=True), 'Devops',
             np.where(df.Position.str.contains('Tech', regex=True), 'Lead',
             np.where(df.Position.str.contains('Designer', regex=True), 'Design',
                      'QA')))))))
    return df
def anal_position(df , work_or_seniority):
    grouped_n_avg_df = group_n_add_cost_avg(add_seniority(df), [work_or_seniority])
    grouped_n_avg_df['Time_p%_4m_total'] = round(100 * grouped_n_avg_df.Time_Decimal / sum(grouped_n_avg_df.Time_Decimal),2)
    grouped_n_avg_df['Cost_p%_4m_total'] =  round(100 * grouped_n_avg_df.Task_Cost / sum(grouped_n_avg_df.Task_Cost),2)
    return grouped_n_avg_df
```
